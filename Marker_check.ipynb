```python
import pandas as pd
import glob
import os

# ---フォルダのパスを指定---
folder_path = "WearableDevices"

# ---指定フォルダ内の全CSVファイルのパスを取得---
# *.csv ワイルドカードを使ってCSVファイルのみをリストアップします
file_paths = glob.glob(os.path.join(folder_path, "*.csv"))

# ---各ファイルに対して処理を実行---
for path in file_paths:
    try:
        # CSVの読み込み
        df = pd.read_csv(path, header=0)
        
        # 'Marker'列が存在する場合のみ処理
        if 'Marker' in df.columns:
            # 'Marker'がNaNではないデータのみ抽出
            df_clean = df[df['Marker'].notna()]
            
            # ファイル名と集計結果を表示
            file_name = os.path.basename(path) # パスからファイル名だけを取り出す
            print(f"--- File: {file_name} ---")
            print(df_clean['Marker'].value_counts())
            print("\n") # 見やすくするために改行を入れる
            
        else:
            print(f"Warning: 'Marker' column not found in {os.path.basename(path)}")
            
    except Exception as e:
        print(f"Error reading {os.path.basename(path)}: {e}")
```
